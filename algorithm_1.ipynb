1. Подберите скорость обучения (eta) и количество итераций

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
n = X.shape[0]

eta = 1e-2 
n_iter = 100

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

```python
n = X.shape[0]
eta_n = 1e-3
n_iter = 200
##пройдёмся циклом по шагу от 0.001 до 0.1, подберём сначала оптимальный шаг
while eta_n < 1e-1:
    eta = eta_n
    W = np.array([1, 0.5])
    print(f'Number of objects = {n} \
    \nLearning rate = {eta} \
    \nInitial weights = {W} \n')
    
    
    for i in range(n_iter):
        y_pred = np.dot(X, W)
        err = calc_mse(y, y_pred)
        for k in range(W.shape[0]):
            W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
        if i % 10 == 0:
            eta /= 1.1
            print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
    eta_n+=1e-3
```

Оптимальный шаг получился в диапазоне от 0,06 до 0,09, поисследуем итерации

```python
n = X.shape[0]
eta_n = 0.06
n_iter = 200

while eta_n < 0.09:
    eta = eta_n
    W = np.array([1, 0.5])
    print(f'Number of objects = {n} \
    \nLearning rate = {eta} \
    \nInitial weights = {W} \n')
##поисследуем влияние итераций на ошибку    
    for n_iter in range(200, 700, 100):
        print(f'количество итераций {n_iter}')
        for i in range(n_iter):
            y_pred = np.dot(X, W)
            err = calc_mse(y, y_pred)
            for k in range(W.shape[0]):
                W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
            if i % 10 == 0:
                eta /= 1.1
                print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
    eta_n+=1e-3
```
